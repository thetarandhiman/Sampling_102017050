```python
import numpy as np
import pandas as pd
```

Load the dataset

```python
df = pd.read_csv('Creditcard_data.csv')
print(df.shape)

df.head()
```

```python
# Drop the Time column as it is not required for the analysis
df.drop('Time', axis=1, inplace=True)
df.head()
```

```python
df['Class'].value_counts()
# The dataset is imbalance dataset because the number of zero class is much more than the number of one class. so we have to convert it into balanced dataset first.
```

Balancing the dataset

```python
X = df.drop('Class', axis=1)
Y = df['Class']
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
test_df = pd.concat([X_test, y_test], axis=1)
```

```python
# Balancing the dataset using overfitting technique i.e. SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=0)
X, Y = sm.fit_resample(X_train, y_train)
X.shape, Y.shape
```

```python
Y.value_counts()
```

```python
# Generating the balanced dataset
df = pd.concat([X, Y], axis=1)
```

```python
df.head()
```

```python
df.tail()
```

## Sampling Techniques & Applying of models

Simple Random Sampling

n = (Z^2 * p * (1-p)) / E^2


```python
import math
import random

n = (1.96**2 * 0.5 * 0.5) / (0.05**2)
n = math.ceil(n)
print(n)

X_sample = X.sample(n=n, random_state=42)
print(X_sample.shape)

y_sample = Y[X_sample.index]
print(y_sample.shape)
```

```python
X_sample.head()
```

```python
y_sample.head()
```

```python
# Applying various models on this random sample
# logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

lr = LogisticRegression()
lr.fit(X_sample, y_sample)
y_pred = lr.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_sample, y_sample)
y_pred = nb.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_sample, y_sample)
y_pred = rf.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_sample, y_sample)
y_pred = dt.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_sample, y_sample)
y_pred = knn.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

Stratified Sampling

n = (Z^2 * p * (1-p)) / (E/S)^2

```python
# Stratified Sampling
df = pd.concat([X, Y], axis=1)
# df['Class'].value_counts()
stratified_df = df.groupby('Class').apply(lambda x: x.sample(frac=0.05, random_state=42))
print(stratified_df['Class'].value_counts())

X_stratified = stratified_df.drop('Class', axis=1)
y_stratified = stratified_df['Class']
```

```python
# decision tree

dt = DecisionTreeClassifier()
dt.fit(X_stratified, y_stratified)
y_pred = dt.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

Cluster Sampling

n = (Z^2 * p * (1-p)) / (E/C)^2


```python
# Cluster Sampling

# defining variables

length = len(df)
groups = 5
size = length / groups
print(size)
```

```python
# group list creation

group_list = []
group_id = 0
size_count = 0

for i in range(length):
    if size_count < size:
        group_list.append(group_id)
        size_count += 1
    else:
        group_id += 1
        group_list.append(group_id)
        size_count = 1

df['group'] = group_list
df.head()
```

```python
# allocating the elements in their respective groups randomly

df['group'] = df['group'].apply(lambda x: random.randint(0, groups-1))
df.head()
```

```python
# randomnly choosing the group

group = random.randint(0, groups-1)
print(group)
```

```python
# Defining the clustered sample

clustered_df = df[df['group'] == group]
print(clustered_df.shape)

# clustered_df.head()
#Drop the group column
clustered_df = clustered_df.drop('group', axis=1)
clustered_df.head()

X_clustered = clustered_df.drop('Class', axis=1)
y_clustered = clustered_df['Class']

X_clustered.shape, y_clustered.shape
```

```python
# logistic regression

lr = LogisticRegression()
lr.fit(X_clustered, y_clustered)
y_pred = lr.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# naive bayes

nb = GaussianNB()
nb.fit(X_clustered, y_clustered)
y_pred = nb.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# random forest

rf = RandomForestClassifier()
rf.fit(X_clustered, y_clustered)
y_pred = rf.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# decision tree

dt = DecisionTreeClassifier()
dt.fit(X_clustered, y_clustered)
y_pred = dt.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
# knn

knn = KNeighborsClassifier()
knn.fit(X_clustered, y_clustered)
y_pred = knn.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

Systematic Sampling

```python
# Systematic Sampling

# get interval value

n = (1.96**2 * 0.5 * 0.5) / (0.05**2)
# n = math.ceil(n)
length = len(df)
interval = length / n
print(interval)
```

```python
# get the starting index

start = np.random.randint(0, interval)
```

```python
systematic_sample = df.iloc[start::int(interval)] 

systematic_sample.head()
```

```python
X_systematic = systematic_sample.drop('Class', axis=1)
y_systematic = systematic_sample['Class']
```

```python
# decision tree

dt = DecisionTreeClassifier()
dt.fit(X_systematic, y_systematic)
y_pred = dt.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```
